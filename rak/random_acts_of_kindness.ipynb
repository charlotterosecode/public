```python
## Initialize
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

k = pd.read_csv('act_of_kindness_ideas.csv')

D = pd.Series(k.Photo_Name.values,index=k.Photo_ID).to_dict()

def random_act():
    x = len(k)-1
    r = np.random.randint(0,x)
    print(r)
    print(k['Act of Kindness Ideas'].iloc[r])
    print(k['Link to Site for Action'].iloc[r])
    
    
    img = mpimg.imread(D[r])
    imgplot = plt.imshow(img)
    ax = plt.gca()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    plt.show()
    
    return
```

```python
## Run Random Acts of Kindness Function
random_act()
```
